```python
# load standard libraries
import numpy as np   # standard numerics library
import matplotlib.pyplot as plt   # for making plots
from numpy import linalg as LA #Linear algebra library


%matplotlib inline
```

```python
# By hand we have calculated the probability distribution p(x,y), now it us just a job computing the entropy of the different measurement angles!

# Assign individual probabilites on the form pxy
n=100
theta=np.linspace(0.001,np.pi,n)

# Insert probabilites computed in tutorial
p11=np.cos(theta)**2/2
p12=np.sin(theta)**2/2
p21=1/4*(1 + 2*np.cos(theta)*np.sin(theta))
p22=1/4*(1 - 2*np.cos(theta)*np.sin(theta))

# Computing relevant entropies
Hxy=-p11*np.log2(p11) -p12*np.log2(p12) -p21*np.log2(p21) -p22*np.log2(p22)
Hy=-(p11 + p21)*np.log2(p11 + p21) - (p12 + p22)*np.log2(p12 + p22)
Hx=1

# Extract peak index and value
maxIndex=np.argmax(Hx + Hy-Hxy)
maxValue=np.amax(Hx + Hy-Hxy)


plt.plot(theta,Hxy, label=r'$H(X,Y)$')
plt.plot(theta,Hy,label=r'$H(Y)$')
plt.plot(theta,Hy-Hxy,label=r'$H(Y)-H(X,Y)$')
plt.xlabel(r'$\theta$')
plt.legend()
plt.show()

plt.plot(theta,Hx + Hy - Hxy,label=r'$H(X:Y)$')
plt.legend()
plt.show()

print(f'We have extracted the maximal accessible information at {theta[maxIndex]}, where the mutual information is {maxValue}.')
```
